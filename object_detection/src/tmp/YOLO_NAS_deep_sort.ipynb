```python


def get_prediction(image_in, pipeline):
  ''' Obtains DetectionPrediction object from a single input RGB image
  '''
  # Preprocess
  preprocessed_image, processing_metadata = pipeline.image_processor.preprocess_image(image=image_in.copy())

  # Predict
  with torch.no_grad():
    # IF CUDA AVAILABLE
    if torch.cuda.is_available():
        torch_input = torch.Tensor(preprocessed_image).unsqueeze(0).to('cuda')
    else:
        torch_input = torch.Tensor(preprocessed_image).unsqueeze(0)
    model_output = model(torch_input)
    prediction = pipeline._decode_model_output(model_output, model_input=torch_input)

  # Postprocess
  return pipeline.image_processor.postprocess_predictions(predictions=prediction[0], metadata=processing_metadata)
```

```python
# Helper function for bounding box colors

import colorsys

def get_color(number):
    """ Converts an integer number to a color """
    # change these however you want to
    hue = number*30 % 180
    saturation = number*103 % 256
    value = number*50 % 256

    # expects normalized values
    color = colorsys.hsv_to_rgb(hue/179, saturation/255, value/255)

    return [int(c*255) for c in color]
```

```python
def get_video_resolution(video_path):
    # Open the video file
    video = cv2.VideoCapture(video_path)

    # Check if the video was successfully opened
    if not video.isOpened():
        print("Error: Could not open video.")
        return None

    # Get width and height
    width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))

    # Release the video object
    video.release()

    return (width, height)
```

```python


# Initialize the tracker with specified parameters:
# det_thresh: threshold for detections to be considered valid
# max_age: maximum number of frames to keep a track alive without updates
# min_hits: minimum number of hits to consider a track as confirmed
tracker = ocsort.OCSort(det_thresh=0.30, max_age=10, min_hits=2)

cap = cv2.VideoCapture(video_path)  # Open the video file

# Check if the video was opened successfully
if not cap.isOpened():
    print("Error opening video file")

frames = []  # List to store frames for later use (e.g., saving video)
i = 0        # Frame counter
counter, fps, elapsed = 0, 0, 0  # Variables for FPS calculation
start_time = time.perf_counter()  # Start time for FPS calculation

while cap.isOpened():  # Loop until video is fully read

    # Read each video frame (time to read frame ~ 0.006 sec)
    ret, frame = cap.read()

    if ret:  # If a frame is successfully read

        # Convert frame to RGB (OpenCV loads images in BGR by default)
        og_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Resize the frame for faster inference (reduce dimensions by half)
        frame = cv2.resize(og_frame, (w2, h2), interpolation=cv2.INTER_LINEAR)

        # Perform object detection on the resized frame using YOLO model
        # pred: contains predicted bounding boxes, confidence, and classes
        pred = model.predict(frame, iou=0.35, conf=0.30)
        pred = pred.prediction  # Extract predictions

        print(pred)

        # Combine bounding box coordinates (x1, y1, x2, y2) with confidence scores
        xyxyc = np.hstack((pred.bboxes_xyxy, np.c_[pred.confidence]))

        # Update the tracker with the detected objects
        # tracks: list of tracked objects with updated states
        tracks = tracker.update(xyxyc, img_info, img_size)

        # Draw tracking information on the original frame
        for track in tracker.trackers:
            #Each track is one tracked object
            time.sleep(1)
            print("state:", track.get_state())  # Get the current state of the track (position)
            print("id:", track.id)
            # print("velocity:", track.velocity)
            # print("kf:", track.kf)
            # print("delta_t:", track.delta_t)
            # print("observation:", track.observations)
            track_id = track.id  # Unique ID of the tracked object
            hits = track.hits  # Number of frames the object has been tracked
            color = get_color(track_id * 15)  # Get color for drawing based on ID
            x1, y1, x2, y2 = np.round(track.get_state()).astype(int).squeeze()  # Bounding box coordinates

            # Draw bounding box and track ID on the frame
            cv2.rectangle(og_frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(og_frame, f"{track_id}-{hits}", (x1 + 10, y1 - 5),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 1, cv2.LINE_AA)

        # Update FPS calculation
        current_time = time.perf_counter()
        elapsed = current_time - start_time  # Time since last FPS update
        counter += 1  # Frame counter for FPS calculation
        if elapsed > 1:  # Update FPS every second
            fps = counter / elapsed  # Calculate FPS
            counter = 0  # Reset frame counter
            start_time = current_time  # Reset start time

        # Display FPS on the frame
        cv2.putText(og_frame, f"FPS: {np.round(fps, 2)}", (10, h - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)

        # Append the frame with drawings to the list
        frames.append(og_frame)

        # TEMP for debug: Uncomment to stop after a few frames
        # if i == 10:
        #     break
        # else:
        #     i += 1

    else:  # Break the loop if no more frames are available
        break

# Release the video capture object and free resources
cap.release()
del cap
```

```python
tracker = ocsort.OCSort(det_thresh=0.30, max_age=10, min_hits=2)
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error opening video file")

frames = []
i = 0
counter, fps, elapsed = 0, 0, 0
start_time = time.perf_counter()

# Dictionary to store track IDs and their corresponding class labels
track_class_dict = {}

while cap.isOpened():

    ret, frame = cap.read()
    if ret:
        og_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame = cv2.resize(og_frame, (w2, h2), interpolation=cv2.INTER_LINEAR)

        # Perform inference to get detections
        pred = model.predict(frame, iou=0.35, conf=0.30)
        pred = pred.prediction

        # Extract bounding boxes and classes from predictions
        xyxyc = np.hstack((pred.bboxes_xyxy, np.c_[pred.confidence]))

        # Update tracker
        tracks = tracker.update(xyxyc, img_info, img_size)

        # Match tracks with predictions
        for idx, track in enumerate(tracker.trackers):
            track_id = track.id  # Unique ID of the tracked object
            track_bbox = np.round(track.get_state()).astype(int).squeeze()  # Bounding box coordinates

            # Get the corresponding class from predictions using the same index
            # object_class = pred.labels[idx]

            # Add to dictionary if not already present
            if track_id not in track_class_dict:
                if idx < len(pred.labels):
                    object_class = pred.labels[idx]
                    track_class_dict[track_id] = object_class

            # Draw the track and ID
            hits = track.hits
            color = get_color(track_id * 15)
            x1, y1, x2, y2 = track_bbox
            cv2.rectangle(og_frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(og_frame, f"{track_id}-{hits}", (x1 + 10, y1 - 5),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 1, cv2.LINE_AA)

        # Update FPS calculation and display on frame
        current_time = time.perf_counter()
        elapsed = current_time - start_time
        counter += 1
        if elapsed > 1:
            fps = counter / elapsed
            counter = 0
            start_time = current_time
        cv2.putText(og_frame, f"FPS: {np.round(fps, 2)}", (10, h - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)

        frames.append(og_frame)

    else:
        break

cap.release()
del cap
```

```python
def save_numpy_as_video(frames, video_savepath, video_fps, frame_size):
    """
    Save a list of numpy arrays as an MP4 video.

    Parameters:
    - frames (list of np.array): List of frames where each frame is a numpy array (H, W, 3).
    - video_savepath (str): Path where the video will be saved.
    - video_fps (int): Frames per second of the output video.
    - frame_size (tuple): Size of each frame as (width, height).

    Returns:
    - None
    """

    # Initialize the VideoWriter with the MP4V codec
    out = cv2.VideoWriter(
        video_savepath,
        cv2.VideoWriter_fourcc(*'mp4v'),  # Use 'mp4v' for .mp4 output
        video_fps,
        frame_size
    )

    # Check if the VideoWriter object was initialized correctly
    if not out.isOpened():
        print("Error: Could not open video writer.")
        return

    # Loop through each frame
    for frame in frames:
        # Ensure frame is the correct size and format
        frame = cv2.resize(frame, frame_size)  # Resize to match the frame size
        if frame.shape[2] == 3:
            frame = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)  # Convert RGB to BGR if needed

        # Write the frame to the video
        out.write(frame)

    # Release the VideoWriter
    out.release()
    print(f"Video saved successfully at {video_savepath}")


```
